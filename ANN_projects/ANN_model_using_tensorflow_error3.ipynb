```python
# Consider the impact of training for more or less epochs.
```

```python
import tensorflow as tf
```

```python
mnist=tf.keras.datasets.mnist
```

```python
(training_images, training_labels), (testing_images, testing_labels)=mnist.load_data()
```

```python
training_images=training_images/255
testing_images=testing_images/255
```

```python
model=tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                 tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                 tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
model.compile(optimizer="Adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(training_images, training_labels, epochs=1)
```

```python
classification=model.predict(testing_images)
```

```python
print(classification[123])
```

```python
print(testing_labels[123])
```

```python
model.fit(training_images, training_labels, epochs=15)
```

```python
# Try 15 epochs -- we'll probably get a model with a much better loss than the one with 5.
```

```python
for i in range(0, 100):
    print("--------------------------------{} image-----------------------------".format(i))
    print(classification[i])
```

```python
for j in range(0, 100):
    print("-------------{} label-------------".format(j))
    print(testing_labels[j])
```

```python
# it can be seen almost our model is accurate and same image labels are predicted by our model as original in testing image.
```

```python
model.fit(training_images, training_labels, epochs=30)
```

```python
# Try 30 epochs -- you might see the loss value stops decreasing, and sometimes increases. This is a side 
#effect of something called 'overfitting' which you can learn about [somewhere] and it's something you need 
#to keep an eye out for when training neural networks. There's no point in wasting your time training if you 
#aren't improving your loss, right!
```

```python
for i in range(0, 100):
    print("----------------------{} image prediction---------------------".format(i))
    print(classification[i])
```
