```python
# how do we train a neural network to do above task.
# Using data, by feeding it with set of x and set of y. It should be able to figure out the relationship
# between x and y.
```

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# we imported tensorflow and calling it tf for ease of use.
# then imported a library numpy which helps us to represent our data as lists easily and quickly.
# and at last imported framework for defining a neural network as a set of sequential layers is called keras.
```

```python
model=tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# create a simple neural network having layer 1 and that layer has 1 neurons.
```

```python
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
# now we need to compile our neural network and for doing we need to specify two function, a loss and optimizer.
# when our computer tries to learn it makes a guess may be y=10x+10 instead of original y=2x-1
# so here Loss function measures guessed answers against known correct answers and measures how well or how bad it is.
# Optimizer function used to make another guess. Based on how loss function went, it will try to minimise the Loss.
# may be at this point it will come up y=5x+5 but still it's pretty bad result compare to original.
# it will repeat till the number of EPOCHS.
# But here how tell it to use mean squared error for loss and STOCHASTIC GRADIENT DESCENT for optimizer.
# it uses big math don't need to know but only how it works.
```

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# now we will feed data into it by taking 6'x and 6'y and here we can observe the realtionship between x and y
# would be y=2x-1
# Numpy' provides lots of array type data structures that are a defacto standard way of doing it.
```

```python
model.fit(xs, ys, epochs=1500)
```

```python
# next is to train the neural network where it learns the relationship between x and y is in the model.fit call.
# this is where it will go through the loop, make guess compare with original and measures how good or bad it is
# then make another guess using optimizer for minimise the loss and it will keeps doing untill EPOCHS
```

```python
print(model.predict([10.0]))
```

```python
# now we have trained model can be used to predict the value of unknown x.
```
